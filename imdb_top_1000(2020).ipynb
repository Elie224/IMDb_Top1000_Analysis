```python
# 5 Nettoyer la variable 'Runtime'. Créer une variable qualitative dont les modalités seront 'Short', 'Medium', Long', 'VeryLong' correspondant à des durées plus petites que 80, entre 80 et 120, entre 120 et 180 et plus grandes que 180. Étudier la variable obtenue.
```

```python
data['Runtime_clean'] = data['Runtime'].str.replace(' min', '', regex=False).astype(int)
```

```python
data['Runtime_cat'] = pd.cut(
    data['Runtime_clean'],
    bins=[0, 80, 120, 180, float('inf')],
    labels=['Short', 'Medium', 'Long', 'VeryLong'],
    right=False
)
```

```python
data['Runtime_cat'].value_counts()
```

```python
data['Runtime_cat'].value_counts(normalize=True) * 100
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.countplot(data=data, x='Runtime_cat', order=['Short','Medium','Long','VeryLong'])
plt.title("Répartition des catégories de durée")
plt.show()
```

```python
# Créer une variable qualitative dont les modalités seront '+', '++', '+++' correspondant à des notes plus petites que 8, entre 8 et 8,5 et plus grandes que 8,5. Étudier la variable obtenue
```

```python
data['IMDB_Rating'] = data['IMDB_Rating'].astype(float)
```

```python
data['Rating_cat'] = pd.cut(
    data['IMDB_Rating'],
    bins=[0, 8, 8.5, float('inf')],
    labels=['+', '++', '+++'],
    right=False
)
```

```python
data['Rating_cat'].value_counts()
```

```python
data['Rating_cat'].value_counts(normalize=True) * 100
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.countplot(data=data, x='Rating_cat', order=['+', '++', '+++'])
plt.title("Répartition des catégories de notes IMDB")
plt.show()
```

```python
# Étudier conjointement les variables obtenues aux questions 5 et 6 (représentations graphiques, table de contingence et test du Khi-deux).
```

```python
ct = pd.crosstab(data['Runtime_cat'], data['Rating_cat'])
ct
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(ct, annot=True, fmt='d', cmap='Blues')
plt.title("Table de contingence Runtime_cat × Rating_cat")
plt.show()
```

```python
ct.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title("Répartition des notes selon la durée du film")
plt.xlabel("Catégorie de durée")
plt.ylabel("Nombre de films")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=data, x='Runtime_cat', hue='Rating_cat',
              order=['Short','Medium','Long','VeryLong'],
              hue_order=['+','++','+++'])
plt.title("Notes IMDB selon la durée du film")
plt.show()
```

```python
from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(ct)

print("Chi2 :", chi2)
print("p-value :", p)
print("Degrés de liberté :", dof)
print("Tableau des effectifs attendus :")
print(expected)
```

```python
""" Interprétation du test du Khi‑deux
Hypothèses du test
H₀ (indépendance) : la durée du film (Runtime_cat) et la note IMDB (Rating_cat) sont indépendantes.
H₁ (dépendance) : les deux variables sont associées.
 Analyse de la p‑value
Ta p‑value = 7.87 × 10⁻⁷, donc :
Elle est très largement inférieure à 0.05
On rejette H₀
Il existe une association significative entre la durée du film et la note IMDB
 Conclusion : la durée d’un film influence (ou est associée à) sa note IMDB. """
```
